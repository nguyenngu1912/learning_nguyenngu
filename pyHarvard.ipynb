```python
# Problem set 0 # 5 (Cs50 Harvard)
def main():
    dollars = dollars_to_float(input("How much was the meal? "))
    percent = percent_to_float(input("What percentage would you like to tip? "))
    tip = dollars * percent
    print(f"Leave ${tip:.2f}")


def dollars_to_float(d):
    d = float(d.replace("$",""))
    return d

def percent_to_float(p):
    p = float(p.replace("%","")) 
    return float (p/100)

main()
```

```python
# Problem set 1 # 3 (Cs50 Harvard)
x = input("File name: ").replace(" ","").lower().strip()

if x[x.find(".") + 1:] == "jpg" :
    print("image/jpg")
elif x[x.find(".") + 1:] == "gif" :
    print("image/gif")
elif x[x.find(".") + 1:] == "jpeg" :
    print("image/jpeg")
elif x[x.find(".") + 1:] == "png" :
    print("image/png")    
elif x[x.find(".") + 1:] == "pdf" :
    print("application/pdf")    
elif x[x.find(".") + 1:] == "txt" :
    print("text/plain")    
elif x[x.find(".") + 1:] == "zip" :
    print("application/x-zip-compressed")    
else:
    print("application/octet-stream ")    

```

```python
# Problem set 1 # 5 (Cs50 Harvard)
# breakfast 7:00 to 8:00, lunch: 12:00 and 13:00, dinner: 18:00 and 19:00
# challenge: support a.m. and p.m. (12-hour times)
def main():
    x = input("What time is it? ").strip().replace(" ","")
    time = convert(x)
    if time == "invalid minutes":
        print("ValueError")
    else:
        if 7 <= round(time,1) <= 8:
            print("breakfast time")
        elif 12 <= round(time,1) <= 13:
            print("lunch time")
        elif 18 <= round(time,1) <= 19:
            print("dinner time") 
        else:
            pass

def convert(time):
    xx = time[-2:]
    if xx == "pm":
        time = time.split(xx)[0]
        hours, minutes = time.split(":")
        hours = int(hours) + 12
    else:
        time = time.split(xx)[0]
        hours, minutes = time.split(":")
    
    if 0 <= int(minutes) <= 60:      
        return int(hours) + float(int(minutes)/60)
    else:
        return "invalid minutes"

if __name__ == "__main__":
    main()

```
